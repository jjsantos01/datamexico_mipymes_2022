```stata
clear
set more off, permanently
* Globales
global dir_main "C:\Users\jjsan\OneDrive\dataton_datamexico_financiero"
global dir_data "${dir_main}\datos\enigh_2020"
global dir_tablas "${dir_main}\datos\tablas_regresiones"

cd "$dir_data"
* Convierte caractéres
unicode encoding set latin1
unicode translate datos_modelo.dta
* Inicia log
capture log close
log using "${dir_tablas}\resultados_modelos_enihg_2020", replace text

* Lee archivo
use "${dir_data}/datos_modelo.dta", clear

* Configura diseño muestral
svyset upm [pweight = factor], strata(est_dis) singleunit(centered)
```

```stata
*  Variables independientes
global INDEP mujer edad tiene_internet tiene_celular tiene_contabilidad ib2.nivelaprob ib1.est_socio ib3.tam_loc i.tipoact tiene_empleados i.ingreso_mensual_rango

*### LOGIT ##*
svy: logit fpago_alt ${INDEP}
estimates store modelo_1
margins, dydx(${INDEP}) vce(unconditional) post
estimates store margins_1
```

```stata
* html
global order_coefs

esttab margins_1 using "${dir_tablas}\efectos_marginales_enigh_2020", ///
 cells(b(fmt(4) star) se(fmt(4)) t(fmt(4)) p(fmt(4)) ci(fmt(4))) starlevels(* 0.1 ** 0.05 *** 0.01) lines label ///
 mlabels("modelo_1") ///
 order(${order_coefs}) nobaselevels replace html


capture log close
```
